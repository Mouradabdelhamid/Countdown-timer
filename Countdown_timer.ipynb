```python
import time
my_time = int(input('Insert the countdown in seconds here please : '))
def countdown(t):
  while t > 0:
    mins , secs = divmod(t, 60)
    timeformat = " {0}, {1}".format(mins,secs)
    print('\r', timeformat) 
    time.sleep(1)
    t = t - 1
  print('Fire in the hole')

countdown(my_time)
```
